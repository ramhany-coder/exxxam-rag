```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("100_Python_Questions_and_Answers.pdf")


documents = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

embedder = HuggingFaceEmbeddings(model = "sentence-transformers/all-MiniLM-L6-v2")
chunker = RecursiveCharacterTextSplitter(chunk_size=1000 , chunk_overlap = 200)
docs = chunker.split_documents(documents=documents)
vbd  = Chroma.from_documents(
    documents=docs,
embedding=embedder,
persist_directory="python_q_emb"
)

```
